```python
import math
import statistics

class ScientificCalculator:
    def add(self, x, y):
        return x + y
    
    def subtract(self, x, y):
        return x - y
    
    def multiply(self, x, y):
        return x * y
    
    def divide(self, x, y):
        if y == 0:
            return "Error: Division by zero"
        else:
            return x / y
    
    def power(self, x, y):
        return x ** y
    
    def square_root(self, x):
        if x < 0:
            return "Error: Cannot take square root of negative number"
        else:
            return math.sqrt(x)
    
    def logarithm(self, x):
        if x <= 0:
            return "Error: Cannot take logarithm of non-positive number"
        else:
            return math.log(x)
    
    def exponential(self, x):
        return math.exp(x)
    
    def sin(self, x):
        return math.sin(x)
    
    def cos(self, x):
        return math.cos(x)
    
    def tan(self, x):
        return math.tan(x)
    
    def arcsin(self, x):
        if x < -1 or x > 1:
            return "Error: Invalid input for arcsine function"
        else:
            return math.asin(x)
    
    def arccos(self, x):
        if x < -1 or x > 1:
            return "Error: Invalid input for arccosine function"
        else:
            return math.acos(x)
    
    def arctan(self, x):
        return math.atan(x)
    
    def mean(self, data):
        return statistics.mean(data)
    
    def median(self, data):
        return statistics.median(data)
    
    def mode(self, data):
        try:
            return statistics.mode(data)
        except statistics.StatisticsError:
            return "Error: No unique mode found"
```
